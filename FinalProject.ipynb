```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
```

### Temporary title, just reading and combining some data sets..

```python
# Concatenating metro bike share data

bike_data = pd.DataFrame()

files_path = os.path.join('data', 'metro-bike-share-trips-****-**.csv')
files = glob.glob(files_path)

for file in files:
    csv = pd.read_csv(file)
    bike_data = pd.concat([bike_data, csv], sort=True)

# Converting time data to datetime format, takes some time..

bike_data['start_time'] = pd.to_datetime(bike_data['start_time'])
bike_data['end_time'] = pd.to_datetime(bike_data['end_time'])


bike_data.head()
```

```python
# Reading in collision data

collisions_path = os.path.join('data', 'Traffic_Collision_Data_from_2010_to_Present.csv')
collisions = pd.read_csv(collisions_path)

# Converting to datetime and keeping rows within bike_data timeframe

collisions['Date Occurred'] = pd.to_datetime(collisions['Date Occurred'])
collisions = collisions[collisions['Date Occurred'].dt.date >= bike_data['start_time'].dt.date.min()]
collisions = collisions.reset_index()

collisions.head()
```

```python
# Reading in traffic count data

traffic_path = os.path.join('data', 'Traffic_Count_Data.csv')
traffic = pd.read_csv(traffic_path)

# Converting to datetime and keeping rows within bike_date timeframe

traffic['Count Date'] = pd.to_datetime(traffic['Count Date'])
traffic = traffic[traffic['Count Date'].dt.date >= bike_data['start_time'].dt.date.min()]
traffic = traffic.reset_index()

traffic.head()
```
